### 1. Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)

Before proceeding. let's state the null and not-null hypotesys

- Null Hypothesis(H0): Both average MEDV are the same
- Alternative Hypothesis(H1): Both average MEDV are NOT the same

```python
var_1 = boston_df[boston_df["CHAS"] == 0]["MEDV"]
var_2 = boston_df[boston_df["CHAS"] == 1]["MEDV"]
scipy.stats.ttest_ind(var_1,var_2,axis=0,equal_var=True)
```

We see that p-value is **more** than alpha value of 0.05 therefore **we failed to reject null hypothesis since there isn't statistical significance.**

### 2. Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)

```python
boston_df.loc[(boston_df["AGE"] <= 35),'age_group'] = '35 years and younger'
boston_df.loc[(boston_df["AGE"] > 35) & (boston_df["AGE"]<70),'age_group'] = 'between 35 and 70 years'
boston_df.loc[(boston_df["AGE"] >= 70),'age_group'] = '70 years and older'
```

- Null Hypothesis(H0): the three population means are the same
- Alternative Hypothesis(H1): One of the means is different

```python
low  = boston_df[boston_df["age_group"] == '35 years and younger']["MEDV"]
mid  = boston_df[boston_df["age_group"] == 'between 35 and 70 years']["MEDV"]
high = boston_df[boston_df["age_group"] == '70 years and older']["MEDV"]
```

```python
f_stats, p_value = scipy.stats.f_oneway(low,mid,high,axis=0)
print(f"F-Statistic={round(f_stats, 3)}, P-value={round(p_value,15)}")
```

Since p-value is less than alpha value of 0.05, we to reject null hypothesis because we see the statistical significance

### 3. Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town? (Pearson Correlation)

Before doing the test, let's state the Hypotesis with the null one being the so called "status quo".

- **Null Hypothesis(H0)**: there is no relationship between NOX variable and INDUS variable
- **Alternative Hypothesis(H1)**: NOX and INDUS are correlated with each other

```python
pearson,p_value = scipy.stats.pearsonr(boston_df["NOX"], boston_df["INDUS"])
print(f"Pearson Coefficient value={round(pearson, 6)}, P-value={round(p_value, 30)}")
```

P-value < 0.05, we must reject the Null hypothesis and conclude that there exists a Pearson relationship between Nitric Oxide and non-retail business acres per town.
